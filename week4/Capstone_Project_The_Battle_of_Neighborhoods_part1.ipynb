## Data Extraction and Data Wrangling

### Import necessary Libraries

```python
from geopy.geocoders import Nominatim
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
source = requests.get('https://www.heraldo.com.mx/jalisco/guadalajara/guadalajara/').text
soup = BeautifulSoup(source, 'html5lib')
```

### Extract neighboards

```python
colonias = soup.find(id="elementos")
#temp = soup.find_all("div", {"class": "fila_non"})
temp = colonias.find_all('div')

for i in temp:
    j = i.text.strip()
    if len(j) > 1:
        print(j)
```

### Extract ZIP code

```python
zip_code = soup.find(id="codigos_postales")
temp_zip = zip_code.find_all('div')

zip_list = []
for i in temp_zip:
    j = i.text.strip()
    if j.startswith('44'):
        zip_list.append(j)
```

### Extract Geospatial Coordinates

```python
#print(location.address)
geolocator = Nominatim(user_agent="my_app")

longitud_list = []
latitud_list = []
for pc in zip_list:
    location = geolocator.geocode("{0}, Guadalajara, Jalisco, México".format(pc))
    longitud_list.append(location.longitude)
    latitud_list.append(location.latitude)
```

### Create a DataFrame

```python
my_dic = {'PC':zip_list, 'Longitud':longitud_list, 'Latitud':latitud_list}
df = pd.DataFrame(my_dic)
df.to_csv('guadalajara_coordinates', index = False)
```
